```python
import requests
from bs4 import BeautifulSoup

# Replace 'url_here' with the URL of the webpage you want to scrape
url = 'https://www.panchangam.org/dailypanchangam.php?date=2012-01-01'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Replace 'div_selector' with the CSS selector for the <div> containing the table
    div_selector = 'table-responsive'

    # Find the <div> element with the specified CSS selector
    target_div = soup.select_one(div_selector)

    if target_div:
        # Find the <table> element within the <div>
        table = target_div.find('table')

        if table:
            # Extract and print table data
            for row in table.find_all('tr'):
                cells = row.find_all(['th', 'td'])
                row_data = [cell.get_text(strip=True) for cell in cells]
                print(row_data)
        else:
            print("Table not found within the div.")
    else:
        print("Target div not found on the page.")
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Define the base URL
base_url = 'https://www.panchangam.org/dailypanchangam.php?date={}'

# Define the start and end years
start_year = 2012
end_year = 2023

# Configure Chrome WebDriver options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # Run Chrome in headless mode (no GUI)
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')



# Initialize Chrome WebDriver with the specified options
driver = webdriver.Chrome(options=chrome_options)

# Loop through the years and scrape data
while start_year <= end_year:
    # Format the date in 'yyyy-mm-dd' format
    formatted_date = f'{start_year}-01-01'

    # Construct the URL with the formatted date
    url = base_url.format(formatted_date)

    try:
        # Navigate to the webpage with the table
        driver.get(url)

        # Wait for the content to load (adjust waiting time as needed)
        wait = WebDriverWait(driver, 10)
        table_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.table-responsive table.table.table-striped.table-condensed')))

        # Get the page source after the content is loaded
        page_source = driver.page_source

        # Parse the page source with BeautifulSoup
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find the table within the div
        table = soup.find('div', class_='table-responsive').find('table', class_='table table-striped table-condensed')
        a=[]

        if table:
            # Extract and print table data
            for row in table.find_all('tr'):
                cells = row.find_all(['th', 'td'])
                row_data = [cell.get_text(strip=True) for cell in cells]
                a.append(row_data)
                print(row_data)
        else:
            print("Table not found within the div.")

    except Exception as e:
        print(f"Error for {formatted_date}: {str(e)}")

    # Move to the next year
    start_year += 1

# Quit the WebDriver
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv

# Define the base URL
base_url = 'https://www.panchangam.org/dailypanchangam.php?date={}'

# Define the start and end years
start_year = 2012  # Start year
end_year = 2023    # End year

# Configure Chrome WebDriver options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # Run Chrome in headless mode (no GUI)
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Initialize Chrome WebDriver with the specified options
driver = webdriver.Chrome(options=chrome_options)

# Define the CSV file and its header
csv_filename = 'panchang_data.csv'
csv_header = ['Date', 'Day', 'Tithi', 'Nakshatra', 'Yoga', 'Karana', 'Calendar']

# Open the CSV file for writing
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)

    # Write the header row
    csv_writer.writerow(csv_header)

    # Loop through the years, months, and days, and scrape data
    while start_year <= end_year:
        current_date = datetime(start_year, 1, 1)  # Start with January 1st of the current year

        while current_date.year == start_year:
            # Format the date in 'yyyy-mm-dd' format
            formatted_date = current_date.strftime('%Y-%m-%d')

            # Construct the URL with the formatted date
            url = base_url.format(formatted_date)

            try:
                # Navigate to the webpage with the table
                driver.get(url)

                # Wait for the content to load (adjust waiting time as needed)
                wait = WebDriverWait(driver, 10)
                table_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.table-responsive table.table.table-striped.table-condensed')))

                # Get the page source after the content is loaded
                page_source = driver.page_source

                # Parse the page source with BeautifulSoup
                soup = BeautifulSoup(page_source, 'html.parser')

                # Find the table within the div
                table = soup.find('div', class_='table-responsive').find('table', class_='table table-striped table-condensed')

                if table:
                    # Extract and print table data
                    row_data = [formatted_date]  # Add the date to the row_data list
                    for row in table.find_all('tr'):
                        cells = row.find_all(['th', 'td'])
                        cell_data = [cell.get_text(strip=True) for cell in cells]
                        row_data.extend(cell_data)  # Extend the row_data list with cell_data

                    # Write the row_data to the CSV file
                    csv_writer.writerow(row_data)
                else:
                    print("Table not found within the div.")

            except Exception as e:
                print(f"Error for {formatted_date}: {str(e)}")

            # Move to the next day
            current_date += timedelta(days=1)

        # Move to the next year
        start_year += 1

# Quit the WebDriver
driver.quit()

print(f"Data has been saved to {csv_filename}")
```
